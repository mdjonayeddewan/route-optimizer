# Distance Matrix Test
Generate a haversine distance matrix using the core distance builder.

```python
import pandas as pd
from src.api.geocoding import LocationPoint
from src.core.distance_builder import build_haversine_matrix

df = pd.read_csv('../data/raw/delivery_points.csv')
points = [LocationPoint(name=row['name'], latitude=row['latitude'], longitude=row['longitude']) for _, row in df.iterrows()]
matrix = build_haversine_matrix(points).distance_matrix_km
pd.DataFrame(matrix, columns=df['name'], index=df['name'])
```
