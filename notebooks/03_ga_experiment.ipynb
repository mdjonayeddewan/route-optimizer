# GA Experiment
Run a quick GA optimization with a synthetic matrix to inspect convergence history.

```python
from models.ga_optimizer import GAConfig, GeneticAlgorithmOptimizer
from models.fitness_function import route_cost, fitness_from_cost

matrix = [
    [0, 10, 12, 8],
    [10, 0, 6, 7],
    [12, 6, 0, 9],
    [8, 7, 9, 0],
]

cfg = GAConfig(40, 100, 0.1, 0.9, 3, 3, 42)
ga = GeneticAlgorithmOptimizer(chromosome_length=4, config=cfg)

def fitness(chromosome):
    cost = route_cost(chromosome, matrix)
    return fitness_from_cost(cost)

result = ga.evolve(fitness)
print('Best chromosome:', result.best_chromosome)
print('Best fitness:', result.best_fitness)
print('History length:', len(result.history))
```
